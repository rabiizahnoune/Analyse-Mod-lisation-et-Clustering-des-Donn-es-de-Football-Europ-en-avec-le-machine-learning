```python
table_match = pd.read_sql('select * from match',conn)
table_player = pd.read_sql('select * from player',conn)
table_player_attribute = pd.read_sql('select * from Player_Attributes',conn)
table_league = pd.read_sql('select * from League',conn)
table_country = pd.read_sql('select * from country',conn)
table_team = pd.read_sql('select * from team',conn)
table_team_attribute = pd.read_sql('select * from Team_Attributes',conn)

```

```python
table_team_attribute
```

```python
table_match_merger = (table_match 
                     
    
    .merge(table_country, left_on='country_id', right_on='id', how='left')
    .merge(table_league, left_on='league_id', right_on='id', how='left')
    .merge(table_team, left_on='home_team_api_id', right_on='team_api_id', how='left', suffixes=('_home', '_away'))
    .merge(table_player, left_on='home_player_1', right_on='player_api_id', how='left', suffixes=('_match', '_player'))

)   
pd.set_option('display.max_columns', None)
```

```python
table_match_merger.head(200)
```

```python
# Calcul des victoires pour chaque équipe dans chaque ligue
table_match_merger['winner'] = table_match_merger.apply(
    lambda row: row['home_team_api_id'] if row['home_team_goal'] > row['away_team_goal']
    else (row['away_team_api_id'] if row['away_team_goal'] > row['home_team_goal'] else None),
    axis=1
)
table_match_merger['date'] = pd.to_datetime(table_match_merger['date'])

# Extraction de l'année
table_match_merger['year'] = table_match_merger['date'].dt.year

# Filtrer les matchs gagnés
victories = table_match_merger.groupby([ 'winner','year','league_id']).size().reset_index(name='wins')
x = victories.loc[victories.groupby(['year'])['wins'].idxmax()]


# Identifier l'équipe dominante par ligue
#dominant_teams = victories.loc[victories.groupby('league_id')['wins'].idxmax()]
dominance_teams_per_league = pd.merge(victories, table_team,right_on='team_api_id',left_on='winner',how='left')
dominance_teams_per_league = dominance_teams_per_league[['team_long_name', 'team_short_name','team_api_id','year','wins','league_id']]
dominance_teams_per_league
```

```python
x
```

```python
# Identifier les équipes perdantes
table_match_merger['loser'] = table_match_merger.apply(
    lambda row: row['home_team_api_id'] if row['home_team_goal'] < row['away_team_goal']
    else (row['away_team_api_id'] if row['away_team_goal'] < row['home_team_goal'] else None),
    axis=1
)

# Filtrer les matchs perdus et regrouper par league_id et loser pour compter les défaites
defeats = table_match_merger.groupby(['loser','year','league_id']).size().reset_index(name='defeats')

# Identifier les équipes les plus défaites par ligue
most_defeated_teams = defeats.loc[defeats.groupby(['year'])['defeats'].idxmax()]

# Ajouter les noms des équipes
defeated_teams = pd.merge(defeats, table_team, right_on='team_api_id', left_on='loser', how='left')


# Sélectionner les colonnes pertinentes
defeated_teams_per_league = defeated_teams[['team_long_name', 'team_short_name','team_api_id','year','defeats','league_id']]

# Afficher les résultats
defeated_teams_per_league
```

```python
statistics_teams_per_league = pd.merge(dominance_teams_per_league,defeated_teams_per_league,on=['team_api_id','year'],how='left')
statistics_teams_per_league = statistics_teams_per_league.drop(columns=['team_short_name_x','team_long_name_x','league_id_x'])
statistics_teams_per_league.to_csv('statistics_teams_per_league.csv',index=False)
```
